# Проведемо аналіз у розрізі статі.


```python
# Аналітика в розрізі статі
df.groupby('Gender')['Total'].agg(['mean', 'count', 'min', 'max'])
```

Як бачимо з таблиці вище жінки та чоловіки практично однаково роблять покупки за кількістю. Але при цьому макс/мін та середній чеки у жінок вищі.
Чому?

**Гіпотеза_1**: в якомусь із регіонів дохід на жінку вищий, ніж на чоловіка - почитати оф джерела. При цьому ми можемо подивитися такий самий зріз у розрізі регіонів - можливо десь сильний перекіс за кількістю та середнім чеком.

```python
df.pivot_table(index='City', columns='Gender', values='Invoice ID', aggfunc='count')
```

```python
df.pivot_table(index='City', columns='Gender', values='Total', aggfunc='mean')
```

Загалом сильного перекосу по містах немає.

**Гіпотеза_2** - жінки віддають перевагу товарам у більш дорогій ціновій категорії
