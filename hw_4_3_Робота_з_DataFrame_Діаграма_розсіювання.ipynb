**Висновок:**
У більшості випадків фактична температура майже співпадає з температурою, як вона відчувається.
Фактична температура в межах 15-25 градусів відчувається, як 20-25. Це може свідчити про високу вологість. Так воно і є на графіку нижче


```python
# Створення графіку розсіювання з трьома змінними
plt.figure(figsize=(10, 8))

# Побудова графіку розсіювання з прозорістю точок 0.6
plt.scatter(df['temp'], df['atemp'], c=df['humidity'], cmap='coolwarm', alpha=0.6)

# Додавання кольорової шкали
plt.colorbar(label='Вологість')

# Додавання заголовка та підписів до осей
plt.title('Графік розсіювання між температурою, вологістю і тим, як вона відчувається')
plt.xlabel('Температура (°C)')
plt.ylabel('Температура, як відчувається (°C)')

# Показ графіку
plt.show()
```

Дивним є те, що фактична температура в 25-35 градусів відчувається як 12-13 градусів. Тут треба дивитись на вологість, швидкість вітру. Можливо, залежить від сезону і як людина одяглась

```python
# Створення графіку розсіювання з трьома змінними
plt.figure(figsize=(10, 8))

# Побудова графіку розсіювання з прозорістю точок 0.6
plt.scatter(df['temp'], df['atemp'], c=df['windspeed'], cmap='coolwarm', alpha=0.6)

# Додавання кольорової шкали
plt.colorbar(label='Швидкість вітру')

# Додавання заголовка та підписів до осей
plt.title('Графік розсіювання між температурою, вологістю і тим, як вона відчувається')
plt.xlabel('Температура (°C)')
plt.ylabel('Температура, як відчувається (°C)')

# Показ графіку
plt.show()
```

В розрізі сезонів взагалі не дуже зрозуміло

```python
# Створення графіку розсіювання з різними кольорами для кожного сезону
plt.figure(figsize=(10, 8))

# Розділення даних за сезонами і побудова графіків розсіювання
for season in range(1, 5):
    season_data = df[df['season'] == season]
    plt.scatter(season_data['temp'], season_data['atemp'], label=f'Сезон {season}')

# Додавання легенди та заголовка
plt.legend()
plt.title('Залежність відчуття температури від фактичної температури за сезонами')
plt.xlabel('Температура (°C)')
plt.ylabel('Відчуття температури (°C)')

# Показ графіку
plt.show()
```
